# Bi-variate Analysis

Two variables are involved in bivariate data. Bivariate analysis is concerned with causes and relationships, and the goal is to determine the relationship between the two variables. 

Comparisons, correlations, causes, and explanations are all part of bivariate data analysis. One of these variables is independent, while the other is dependent, and these variables are frequently plotted on the X and Y axis on the graph for a better understanding of the data. Bivariate analysis can be divided into three categories. they are 
Analysis of two Numerical Variables ,
Analysis of two Categorical Variables ,
Analysis of One Numerical and One Categorical Variable ,

```python
# Checking wheather income and seniority columns are co related with other or not
# Gross income of the household vs Customer seniority in month

#Linear Correlation â€“
# The strength of a linear link between two numerical variables is represented by linear correlation. 
# The value of r denotes the strength of a linear relationship and is always between -1 and 1.
```

```python
import matplotlib.pyplot as plt
import scipy.stats
plt.style.use('ggplot')
slope, intercept, r, p, stderr = scipy.stats.linregress(data["Customer seniority"], data["Gross income of the household"])
best_fit_line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
best_fit_line
```

```python
fig, ax = plt.subplots()
ax.plot(data["Customer seniority"], data["Gross income of the household"], linewidth=0, marker='s', label='Data points')
ax.plot(data["Customer seniority"], intercept + slope * data["Customer seniority"], label=best_fit_line)
ax.set_xlabel('Customer Seniority in Months')
ax.set_ylabel('Income')
ax.legend(facecolor='white')
plt.show()
```

```python
import numpy as np
corr_matrix = np.corrcoef(data["Customer seniority"], data["Gross income of the household"]).round(decimals=2)
corr_matrix
```

```python

fig, ax = plt.subplots()
im = ax.imshow(corr_matrix)
im.set_clim(-1, 1)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Customer seniority', 'income'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Customer seniority', 'income'))
for i in range(2):
    for j in range(2):
        ax.text(j, i, corr_matrix[i, j], ha='center', va='center',
                color='black')
cbar = ax.figure.colorbar(im, ax=ax, format='% .2f')
plt.show()
```

###  Chi-Square Test - Bivariate Analysis

The chi-square test is used for determining the association between categorical variables. It is calculated based on the difference between expected frequencies and the observed frequencies in one or more categories of the frequency table. A probability of zero indicates a complete dependency between two categorical variables and a probability of one indicates that two categorical variables are completely independent. 

```python
# H0 (Null Hypothesis) = The 2 variables to be compared are independent.
# H1 (Alternate Hypothesis) = The 2 variables are dependent.

# p-value obtained after conducting the test is less than 0.05 we reject the Null hypothesis and accept the Alternate hypothesis and 
# if the p-value is greater that 0.05, we accept the Null hypothesis and reject the Alternate hypothesis. 
```

```python
from sklearn.feature_selection import chi2

resultant = pd.DataFrame(data=[(0 for i in range(len(obj_data.columns))) for i in range(len(obj_data.columns))], 
columns=list(obj_data.columns))
resultant.set_index(pd.Index(list(obj_data.columns)), inplace = True)

# Finding p_value for all columns 
for i in list(obj_data.columns):
    for j in list(obj_data.columns):
        if i != j:
            chi2_val, p_val = chi2(np.array(obj_data[i]).reshape(-1, 1), np.array(data[j]).reshape(-1, 1))
            resultant.loc[i,j] = p_val
```

```python
# Plotting a heatmap
import seaborn as sns
fig = plt.figure(figsize=(15,18))
sns.heatmap(resultant, annot=True, cmap='Blues')
plt.title('Chi-Square Test Results')
plt.show()
```

# Multi-variate Analysis

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
preprocessor = Pipeline([("scaler", MinMaxScaler()),("pca", PCA(n_components=2, random_state=42)),])
```

```python
clusterer = Pipeline([("kmeans",
KMeans(
n_clusters=5,
init="k-means++",
n_init=50,
max_iter=500,
random_state=42,),),])
```

```python
pipe = Pipeline([("preprocessor", preprocessor),("clusterer", clusterer)])
pipe.fit(data)
```
